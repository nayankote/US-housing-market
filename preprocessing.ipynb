# DEMAND

mortgage rates, contains weekly data; Source - https://fred.stlouisfed.org/series/MORTGAGE30US

```python
df_morates = pd.read_excel('demand/mortgage_rates.xls',skiprows=10,names=['observation_date','mortgage_rates'])
print(df_morates.describe())
mortgage_dec1999 = 7.95
df_morates.head()
```

```python
# This is weekly data, taking average over month
df_morates['month'] = pd.DatetimeIndex(df_morates['observation_date']).month
df_morates['year'] = pd.DatetimeIndex(df_morates['observation_date']).year
monthly_mortgages = []
for date in month_indices:
    monthly_mortgages.append(np.round(np.average(df_morates.loc[(df_morates['month'] == date.month) & (df_morates['year'] == date.year)]['mortgage_rates']),2))
df_mortgage = pd.DataFrame({"mortgage_rates":monthly_mortgages}, index = month_indices)
print(df_mortgage.head())
df_mortgage['mortgage_rates'].plot()
plt.title("Mortgage Rates")
plt.show()
```

## Full Demand Data

```python
# All demand data 
demand_data = pd.concat([df_hai, df_cons_sent, df_cpi_rent, df_rdi, df_mhi, df_mortgage], axis=1)
demand_data.interpolate(inplace=True)
demand_data['median_household_income'] = np.round(demand_data['median_household_income'],0)
mort_rates = np.insert(demand_data['mortgage_rates'].to_numpy(),0,mortgage_dec1999) # mortgage_dec1999 = 7.95%
demand_data['mortgage_rates_change'] = (mort_rates[1:] - mort_rates[:-1])*100/mort_rates[:-1]
demand_data.head()
```
